```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
train_df = pd.read_csv('train.csv')
train_df.head()
```

```
image_id=[]
cate=[]
for i in range(len(train_df["ImageId_ClassId"])):
  lis = train_df["ImageId_ClassId"][i].split("_")
  print(lis[0])
  image_id.append(lis[0])
  cate.append(lis[1])

train_df["image_id"]=image_id
train_df["category"]=cate
```

```
train_df.drop("ImageId_ClassId" , axis = 1,inplace = True)
```

```
train_df.head()
```

```
df_new = pd.DataFrame(columns=['image_id','0','1','2','3','4'])
```

```
lis = np.unique(train_df["image_id"])
```

```
df_new['image_id'] = lis
```

```
for i in range(len(train_df)):
  cat = train_df["category"][i]
  encP = train_df["EncodedPixels"][i]

  if encP is np.nan:
    continue
  else:
    
    df_new[str(cat)][i]  = 1
    
```

```
df_new.fillna(0,inplace=True)
```

```
df_new['0'] = df_new['4'] + df_new['3'] + df_new['2'] + df_new['1']
```

```
df_new ['0']=df_new['0']*-1 +1
```

```
df_new['target'] = np.argmax(df_new.T["0":].T.values,axis=1)
```

```
df_new['target']=df_new['target'].astype(str)
```

```
df_new.head()
```

```
val_df=df_new[:500]
train__df  = df_new[500:]
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
        dataframe=train__df,
        directory='train_images',
        x_col="image_id",
        y_col="target",
        target_size=(256, 256),
        batch_size=16,
        class_mode='sparse')

validation_generator = test_datagen.flow_from_dataframe(
        dataframe=val_df,
        directory='train_images',
        x_col="image_id",
        y_col="target",
        target_size=(256, 256),
        batch_size=32,
        class_mode='sparse')
```

```
import tensorflow as tf
```

```
from tensorflow.keras.models import Sequential , Model
from tensorflow.keras.layers import Dense , Flatten , BatchNormalization , Conv2D , MaxPooling2D , GlobalMaxPooling3D , GlobalAveragePooling2D , Conv3D , Activation
```

```
from tensorflow.keras.applications import *
```

```
from tensorflow.python.keras.applications.resnet import ResNet101
```

```
from keras import regularizers
```

```
!wget https://github.com/keras-team/keras-applications/''releases/download/densenet/densenet121_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```
model = Sequential()
res =tf.keras.applications.resnet_v2.ResNet152V2(include_top=False, weights=None, input_shape=(256,256,3) ,  pooling='max')
```

```
x=res.output
x = BatchNormalization()(x)
```

```
 x = Dense(1000 , kernel_regularizer=regularizers.l1_l2(0.01), activity_regularizer=regularizers.l2(0.01))(x)
```

```
x = Activation('relu')(x)
x = Dense(500, kernel_regularizer=regularizers.l1_l2(0.01), activity_regularizer=regularizers.l2(0.01))(x)
x = Activation('relu')(x)
predictions = Dense(5, activation='softmax')(x)
model = Model(inputs=res.input, outputs=predictions)
```

```
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```

model.fit_generator(
        train_generator,
        steps_per_epoch=10,
        epochs=20,
        validation_data=validation_generator,
        validation_steps=10)
```

```
model.save('res_net_class152.h5')
```
